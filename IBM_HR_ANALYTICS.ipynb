```python
data = data.drop(["EmployeeCount", "StandardHours"], axis=1)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data.select_dtypes(include='number').corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
data1 =['Age','Education','MonthlyIncome','YearsAtCompany','YearsInCurrentRole','DistanceFromHome','PerformanceRating','TotalWorkingYears','YearsWithCurrManager']
correlation_matrix=data[data1].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
